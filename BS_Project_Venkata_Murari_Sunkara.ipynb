### 1.1 What is the probability that a randomly chosen player would suffer an injury?

### 1.2 What is the probability that a player is a forward or a winger?

```python
Total_Forwards = 94
Total_Wingers = 29
Total_Players = 235

p_forward_or_winger = (Total_Forwards + Total_Wingers)/ Total_Players
print(f"The probability that a player is a forward or a winger is {np.round(p_forward_or_winger,2)*100} %")
```

### 1.3 What is the probability that a randomly chosen player plays in a striker position and has a foot injury?

```python
Striker_injured = 45
Total_Players = 235

p_Striker_Injury = Striker_injured / Total_Players
print(f"The probability that a randomly chosen player plays in a striker position and has a foot injury is {np.round(p_Striker_Injury,2)*100} %")
```

### 1.4 What is the probability that a randomly chosen injured player is a striker?

```python
Striker_injured = 45
Players_Injured = 145

p_injured_striker = Striker_injured/ Players_Injured
print(f"The probability that a randomly chosen injured player is a striker is {np.round(p_injured_striker,2)*100} %")
```

**Import `norm` from `scipy.stats`**

```python
from scipy.stats import norm
```

**Plotting the Normal Distribution**

```python
mu = 5
sigma = 1.5
x = np.linspace(mu-3*sigma, mu+3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.title('Normal Distribution')
plt.show()
```

This plot shows the normal distribution curve with a mean (μ) of 5 and a standard deviation (σ) of 1.5. 

## 2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python
Q_2_1 = norm.cdf(3.17, mu, sigma)
print("Proportion of gunny bags with breaking strength at least 3.17 kg per sq cm:", np.round(Q_2_1,5))
```

```python
mu = 5
sigma = 1.5
x = np.linspace(mu-3*sigma, mu+3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.axvline(3.17, color='r')
plt.fill_between(x, norm.pdf(x, mu, sigma), where= x< 3.17, color='r')
plt.title('Proportion of gunny bags with breaking strength at least 3.17 kg per sq cm')
plt.show()
```

This plot shows the normal distribution and highlights the area under the curve for breaking strengths less than 3.17 kg per sq cm.

## 2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

```python
Q_2_2  = 1- norm.cdf(3.6, mu, sigma)
print("Proportion of gunny bags with breaking strength at least 3.6 kg per sq cm:", np.round(Q_2_2,4))
```

```python
mu = 5
sigma = 1.5
x = np.linspace(mu-3*sigma, mu+3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.axvline(3.6, color='r')
plt.fill_between(x, norm.pdf(x, mu, sigma), where= x >= 3.6, color='g')
plt.title('Proportion of gunny bags with breaking strength at least 3.6 kg per sq cm')
plt.show()
```

This plot shows the normal distribution and highlights the area under the curve for breaking strengths of at least 3.6 kg per sq cm.

## 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
Q_2_3  = norm.cdf(5.5, mu, sigma) - norm.cdf(5, mu, sigma)
print("Proportion of gunny bags have a breaking strength between 5 and 5.5 kg per sq cm:", np.round(Q_2_3,4))
```

```python
mu = 5
sigma = 1.5
x = np.linspace(mu-3*sigma, mu+3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.axvline(5, color='r')
plt.axvline(5.5, color='r')
plt.fill_between(x, norm.pdf(x, mu, sigma), where= ((x >= 5) & (x <= 5.5)), color='b')
plt.title('Proportion of gunny bags have a breaking strength between 5 and 5.5 kg per sq cm')
plt.show()
```

This plot shows the normal distribution and highlights the area under the curve for breaking strengths between 5 and 5.5 kg per sq cm.

## 2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
i = 1 - norm.cdf(7.5, mu, sigma)
j = norm.cdf(3, mu, sigma)
Q_2_4  = i + j
print("Proportion of gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm:", np.round(Q_2_4,4))
```

# <a name='link4'>Problem 3</a>


**Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image, the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);**


**Loading the Dataset**

```python
Zingaro_company.head()
```

## 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

We will test the null hypothesis

>$H_0:\mu= 150$

against the alternate hypothesis

>$H_a:\mu< 150$

#### **One-Sample t-Test for Unpolished Stones**

```python
mu = 150

test_stat, p_Value = stats.ttest_1samp(Zingaro_company['Unpolished '], popmean= mu, alternative= 'less')
print('The test_stat is',np.round(test_stat,5))
print('The p-value is',p_Value)
```

**Checking the level of significance is greater than p-value**

```python
if p_Value < 0.05:
    print(f'As the p-value {p_Value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_Value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

**Insight**

As the p-value is significantly less than the level of significance (0.05), we can **reject the null hypothesis**. Hence, we do have enough evidence to support the claim that the mean hardness of unpolished stones is less than 150. This indicates that the unpolished stones may not be suitable for printing, as their hardness does not meet the required level.

## 3.2 Is the mean hardness of the polished and unpolished stones the same?

We will test the null hypothesis

>$H_0:\mu_1=\mu_2$

against the alternate hypothesis

>$H_a:\mu_1\neq\mu_2$

**Calculating Mean and Standard Deviation for Unpolished and Polished Stones**

```python
print('The mean for Unpolished stone is',Zingaro_company['Unpolished '].mean())
print('The mean for Treated and Polished stone is',Zingaro_company['Treated and Polished'].mean())
print('The std for Unpolished stone is',Zingaro_company['Unpolished '].std())
print('The std for Treated and Polished stone is',Zingaro_company['Treated and Polished'].std())
```

#### **Two-Sample t-Test for Unpolished and Polished Stones**

```python
test_stat, p_value = stats.ttest_ind(Zingaro_company['Unpolished '], Zingaro_company['Treated and Polished'], alternative='two-sided')
print('The test_stat is', np.round(test_stat,5))
print('The p-value is',np.round(p_value,5))
```

**Checking the level of significance is greater than p-value**

```python
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

#### **Insight**

As the p-value (~0.00147) is less than the level of significance (0.05), we can **reject the null hypothesis**. Hence, we do have enough evidence to support the claim that the mean hardness of polished and unpolished stones is different.

## 4.1 How does the hardness of implants vary depending on dentists?

## Hypothesis Testing

### Step 1: Define null and alternative hypotheses

We will test the null hypothesis

>$H_0: $The mean hardness of implants is the same for all dentists.

against the alternate hypothesis

>$H_a: $The mean hardness of implants is different for at least one dentist.

### Step 2: Select Appropriate test

One-way ANOVA is an appropriate test here, provided normality and equality of variance assumptions are verified.

**One-way ANOVA test**

In a one-way ANOVA test, we compare the means from several populations to test if there is any significance difference between them. The results from an ANOVA test are most reliable when the assumptions of normality and equality of variances are satisfied.

* For testing of normality, Shapiro-Wilk’s test is applied to the response variable.

* For equality of variance, the Levene test is applied to the response variable.

### Unstacking the Alloy Data

```python
alloy_1 = Dental_hardness[Dental_hardness['Alloy'] == 1]
alloy_2 = Dental_hardness[Dental_hardness['Alloy'] == 2]
```

## Shapiro-Wilk’s test

We will test the null hypothesis

>$H_0: $The data follows a normal distribution.

against the alternate hypothesis

>$H_a: $The data does not follow a normal distribution.

### **Alloy type 1**

```python
for i in (Dental_hardness['Dentist'].unique()):
    stat, p_value = stats.shapiro(alloy_1[alloy_1['Dentist'] == i]['Response'])
    print('The P-value for Dentist',i,'is', p_value)

    if p_value < 0.05:
        print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
        print("")
    else:
        print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
        print("")
```

- The p-value for Dentist 4 is 0.05333680287003517

**For Alloy Type 1,** all p-values are greater than 0.05, indicating that the data for each dentist is normally distributed. Therefore, we can assume normal distribution for Alloy Type 1.

### **Alloy type 2**

```python
for i in (Dental_hardness['Dentist'].unique()):
    stat, p_value = stats.shapiro(alloy_2[alloy_2['Dentist'] == i]['Response'])
    print('The P-value for Dentist',i,'is', p_value)
    
    if p_value < 0.05:
        print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
        print("")
    else:
        print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
        print("")
```

## Hypothesis Testing

### Step 1: Define null and alternative hypotheses

We will test the null hypothesis

>$H_0: $The mean hardness of implants is the same for all methods.

against the alternate hypothesis

>$H_a: $The mean hardness of implants is different for at least one method.

### Step 2: Select Appropriate test

One-way ANOVA is an appropriate test here, provided normality and equality of variance assumptions are verified.

**One-way ANOVA test**

In a one-way ANOVA test, we compare the means from several populations to test if there is any significance difference between them. The results from an ANOVA test are most reliable when the assumptions of normality and equality of variances are satisfied.

* For testing of normality, Shapiro-Wilk’s test is applied to the response variable.

* For equality of variance, the Levene test is applied to the response variable.

### Shapiro-Wilk’s test

We will test the null hypothesis

>$H_0: $The data follows a normal distribution.

against the alternate hypothesis

>$H_a: $The data does not follow a normal distribution.

### **Calculate the p-value**

## Two -Way ANOVA

#### **Alloy 1**

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
formula = 'Response ~ C(Dentist) + C(Method) + C(Dentist):C(Method)'
model = ols(formula, alloy_1).fit()
aov_table = anova_lm(model)
(aov_table)
```

```python
row_name = 'C(Dentist):C(Method)'
p_value_interaction_1 = aov_table.loc[row_name, 'PR(>F)']

if p_value_interaction_1 < 0.05:
    print(f'As the p-value {p_value_interaction_1} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value_interaction_1} is greater than the level of significance, we fail to reject the null hypothesis.')
```

#### **Conclusion**

- **Method:** Significant effect on hardness (p-value = 0.000284).
- **Interaction (Dentist & Method):** Significant interaction effect on hardness (p-value = 0.006793)
